# Animal Shelter Outcomes - Machine Learning Model

###  Objetivo

Este proyecto es una continuacin del anlisis de datos exploratorio que realic previamente sobre los **factores que afectan las tasas de adopcin** de perros y gatos en refugios de animales.

Su objetivo es crear un modelo de machine learning que sea capaz de **predecir el resultado de adopcin ms probable para cada perro o gato** basado en sus caractersticas.

###  Descripcin del problema

Nuestra **variable objetivo es el tipo de resultado** u `outcome_type` - si el animal es adoptado, devuelto a su dueo, trasladado a otro centro, eutanasiado...

Esta variable tiene mltiples posibles valores categricos (de tipo `string`) y sin relacin ordinal directa.

Por tanto, se tratara de un problema de **clasificacin** multiclase. Adems, las clases no estn balanceadas, ya que hay resultados que son mucho ms frecuentes.

###  Fuente de los datos

Este proyecto est basado en los datos que limpi previamente en mi proyecto de anlisis de datos exploratorio. 

Originalmente, el dataset fue compilado por Jack Daoud a partir de datos del refugio de animales de Austin, Texas [Austin Animal Center](https://www.austintexas.gov/austin-animal-center) y pueden descargarse en [Kaggle](https://www.kaggle.com/datasets/jackdaoud/animal-shelter-analytics). Los datos originales fueron facilitados por el [Ayuntamiento de Austin](https://data.austintexas.gov) y fueron recopilados de 2013 a 2021.

El [siguiente *notebook*](./IV.%20Modelaje.ipynb) tratar con la parte de limpieza de datos y la seleccin de variables independientes (features).
