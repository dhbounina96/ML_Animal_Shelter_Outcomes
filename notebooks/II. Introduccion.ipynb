# Animal Shelter Outcomes - Machine Learning Model

###  Objetivo

Este proyecto es una continuación del análisis de datos exploratorio que realicé previamente sobre los **factores que afectan las tasas de adopción** de perros y gatos en refugios de animales.

Su objetivo es crear un modelo de machine learning que sea capaz de **predecir el resultado de adopción más probable para cada perro o gato** basado en sus características.

###  Descripción del problema

Nuestra **variable objetivo es el tipo de resultado** u `outcome_type` - si el animal es adoptado, devuelto a su dueño, trasladado a otro centro, eutanasiado...

Esta variable tiene múltiples posibles valores categóricos (de tipo `string`) y sin relación ordinal directa.

Por tanto, se trataría de un problema de **clasificación** multiclase. Además, las clases no están balanceadas, ya que hay resultados que son mucho más frecuentes.

###  Fuente de los datos

Este proyecto está basado en los datos que limpié previamente en mi proyecto de análisis de datos exploratorio. 

Originalmente, el dataset fue compilado por Jack Daoud a partir de datos del refugio de animales de Austin, Texas [Austin Animal Center](https://www.austintexas.gov/austin-animal-center) y pueden descargarse en [Kaggle](https://www.kaggle.com/datasets/jackdaoud/animal-shelter-analytics). Los datos originales fueron facilitados por el [Ayuntamiento de Austin](https://data.austintexas.gov) y fueron recopilados de 2013 a 2021.

El [siguiente *notebook*](./IV.%20Modelaje.ipynb) tratará con la parte de limpieza de datos y la selección de variables independientes (features).
